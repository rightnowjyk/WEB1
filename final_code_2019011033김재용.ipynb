```python
 #Status of electric vehicles in South Korea 꺾은선 그래프로 표현
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
row_numbers = [2, 14, 26, 37]
# 특정 행만 추출하여 데이터 프레임 생성
filtered_EV = EV.iloc[row_numbers]

# 필요한 열 선택


# 필요한 열 선택
dates = filtered_EV['기준일']
seoul = filtered_EV['서울']
incheon = filtered_EV['인천']
gyeonggi = filtered_EV['경기']
seoul = filtered_EV['서울']
incheon = filtered_EV['인천']
gyeonggi = filtered_EV['경기']
gangwon = filtered_EV['강원']
chungbuk = filtered_EV['충북']
chungnam = filtered_EV['충남']
daejeon = filtered_EV['대전']
sejong = filtered_EV['세종']
gyeongbuk = filtered_EV['경북']
daegu = filtered_EV['대구']
jeonbuk = filtered_EV['전북']
jeonnam = filtered_EV['전남']
gwangju = filtered_EV['광주']
gyeongnam = filtered_EV['경남']
busan = filtered_EV['부산']
ulsan = filtered_EV['울산']
jeju = filtered_EV['제주']
total = filtered_EV['합계']



# 나머지 열도 필요한 열로 선택하여 변수에 할당합니다.

# 그래프 설정
plt.figure(figsize=(12, 6))
plt.plot(dates, seoul, label='Seoul')
plt.plot(dates, incheon, label='Incheon')
plt.plot(dates, gyeonggi, label='Gyeonggi')
plt.plot(dates, gangwon, label='Gangwon')
plt.plot(dates, chungbuk, label='Chungbuk')
plt.plot(dates, chungnam, label='Chungnam')
plt.plot(dates, daejeon, label='Daejeon')
plt.plot(dates, sejong, label='Sejong')
plt.plot(dates, gyeongbuk, label='Gyeongbuk')
plt.plot(dates, daegu, label='Daegu')
plt.plot(dates, jeonbuk, label='Jeonbuk')
plt.plot(dates, jeonnam, label='Jeonnam')
plt.plot(dates, gwangju, label='Gwangju')
plt.plot(dates, gyeongnam, label='Gyeongnam')
plt.plot(dates, busan, label='Busan')
plt.plot(dates, ulsan, label='Ulsan')
plt.plot(dates, jeju, label='Jeju')
plt.plot(dates, total, label='Total')

plt.gca().invert_xaxis()  
plt.title('Status of electric vehicles in South Korea')
plt.xlabel('Year')
plt.ylabel('EV')
plt.legend(loc='best')
plt.xticks(rotation=45)

plt.show()
```

```python
import pandas as pd

# LG 에너지 솔루션 분기별 매출액 데이터
sales_data = {
    'quartly': ['Q1', 'Q2', 'Q3', 'Q4'],
    'revenue': [43423, 50706, 76482, 85375]
}

# 데이터 프레임 생성
df_lg = pd.DataFrame(sales_data)

print(df_lg)

```

```python
# 막대 그래프 그리기
plt.bar(df_lg['quartly'], df_lg['revenue'])
plt.xlabel('quartly')
plt.ylabel('revenue(단위:억)')
plt.title('LG Energy Solution quartly revenue')

# 그래프 출력
plt.show()
```
